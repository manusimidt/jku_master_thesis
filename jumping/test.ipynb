```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# generate some dummy data
TIME_WINDOW = 1000 # seconds
data1 = np.random.normal(np.linspace(1, 10, num=TIME_WINDOW), 0.5, size=(3, TIME_WINDOW))
data2 = np.random.normal(0, 2, size=(3, int(TIME_WINDOW / 10)))

print(f"Shapes: data1: {data1.shape}, data2: {data2.shape}")
```

```python
# Resize data2 such that we have 1 measurement for every second instead of every 10 second
data2 = np.repeat(data2, 10.5, axis=1)

print(f"Shapes: data1: {data1.shape}, data2: {data2.shape}")
```

```python
diff = data1 - data2
```

```python
plt.plot(diff[0].T)
```
